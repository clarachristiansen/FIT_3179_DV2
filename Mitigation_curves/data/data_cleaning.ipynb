```python
import numpy as np
import pandas as pd
import math
```

```python
data = pd.read_csv('mitigation_curves_1.5C_191203_data.csv')
```

```python
data.head()
```

```python
data.columns[1:]
```

```python
new_data = pd.DataFrame()
for col in data.columns[1:]:
    for i in data.index:
        if not math.isnan(data[col][i]) and data['Year'][i] >= 1950:
            new_row = pd.DataFrame({'Symbol': [col], 'Date': [data['Year'][i]], 'CO2': [data[col][i]]})
            new_data = pd.concat([new_row, new_data], axis=0)
```

```python
new_data.head()
```

```python
new_data.to_csv('mitigation_curves_cleaned.csv')
```
